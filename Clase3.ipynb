```python
#Definimos la primera función
def auditar_ventas(lista_ventas): #La función se llama auditar_ventas y tiene una incógnita que es lista_ventas
  alertas = 0 #Se inicializa en 0 porque es donde vamos a ir acumulando la suma
  for venta in lista_ventas:
    if venta < 5000:
      alertas=alertas + 1
  #print(alertas) #Con esta función se nos imprimirá el nº de alertas
  return alertas #Con esta función se va a guardar el nº de alertas de cada lista_ventas
```

```python
ventas_lunes=[12000,3000,8500]
ventas_martes=[4000, 1000, 25000, 45000]
alertas_lunes = auditar_ventas(ventas_lunes) #En alertas lunes se va aguardar el return de la función (alertas)
alertas_martes = auditar_ventas(ventas_martes)
print(f"El número de alertas del lunes fueron: {alertas_lunes} y el martes de {alertas_martes}")
```

Funciones dentro de numpy

```python
venta_semana = np.array([1200, 1500, 900, 2100, 1800, 800, 2500 ])
#Media
media=np.mean(venta_semana)
maximo=np.max(venta_semana)
minimo=np.min(venta_semana)
desviacion_tipica=np.std(venta_semana)
print(f"La media es {round(media,2)}, el maximo es {maximo}, el minimo es {minimo} y la desviacioón típica es {round(desviacion_tipica,2)}")
```

```python
#Podemos filtrar muy facil
ventas_top=venta_semana[venta_semana>1500]
print(ventas_top)
```
