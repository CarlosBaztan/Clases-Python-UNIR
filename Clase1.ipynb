OPERACIONES BÁSICAS

```python
#Las variables de una celda se ejecutan en cascada (1ª fila, 2ª,...)
Gastos_empresa1=20000
Beneficios_empresa1=Ingresos_empresa1-Gastos_empresa1
print(f"La {empresa1} tiene unos beneficios de {Beneficios_empresa1}")
```

Vamos a definir los diferentes tipos de variables

```python
#Variable tipo string (texto)
cliente="EmpresaXYZ"
#Variable tipo int (entero)
unidades_vendidas=150
#Variable tipo float (decimal)
precio_unitario=24.99
#Variable tipo Boolean (Verdadero/Falso)
cliente_activo=True
```

¿Cómo se el tipo de variable?

```python
print(f"El tipo de variable {cliente} es {type(cliente)}")
print(type(precio_unitario))
print(type(unidades_vendidas))
print(type(cliente_activo))
```

OPERACIONES

```python
Ingresos_totales=unidades_vendidas*precio_unitario
print(f"Los ingresos totales son {Ingresos_totales}")
```

```python
Gastos_totales=15*unidades_vendidas
Beneficios_totales=Ingresos_totales-Gastos_totales
print(f"La {empresa1} ha obtenido {Beneficios_totales} euros de beneficios")
```

```python
#con round, redondeo a los decimales que quiera (en este caso como seria 1498.50, solo coge 1)
print(f"La {empresa1} ha obtenido {round(Beneficios_totales, 2)} euros de beneficios")
```
