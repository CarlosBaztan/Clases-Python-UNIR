Quiero hacer un programa que lea nuestra lista de ingresos y nos avise si hay una venta muy baja (por ejemplo < 5000 eur) para que el departamento de riesgo la analice


```python
lista_ingresos=[12000, 3000, 8500, 1000, 25000]
#Programa que lea nuestra lista:
for ingreso in lista_ingresos:
  #Se nos avise si hay venta menor a 5000 euros
  if ingreso < 5000:
    print(f"Aviso: Caso de {ingreso} euros a analizar")
```

El reto del "contador" y el "acumulador"

```python
lista_ingresos=[12000, 3000, 8500, 1000, 25000]
#Definir las variables que van a guardar la suma y la cuenta. Las vamos a inicializar a 0.
total_ingresos = 0
total_avisos = 0

for ingresos in lista_ingresos:
  total_ingresos = total_ingresos + ingresos
  if ingresos < 5000:
    total_avisos = total_avisos + 1

print(f"El total de ingresos es {total_ingresos}")
print(f"El total de avisos es {total_avisos}")
```
