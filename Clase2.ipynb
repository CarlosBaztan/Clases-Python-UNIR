```python
#Pedimos el dato al usuario
ingresos_texto= input("Por favor, introduce los ingresos del cliente: ")
ingresos_numero= int(ingresos_texto)
```

```python
if ingresos_numero >= 10000:
  print("Es cliente VIP") #si no se cumple, no se ejecuta
elif ingresos_numero >= 5000:
  print("Es cliente bueno")
elif ingresos_numero >= 2000:
  print("Es cliente regular")
else:
  print("No es cliente VIP")
```

```python
ingresos_2= 10000
cliente_2_activo= True #Para definir variables usamos 1 igualdad
if ingresos_2 >= 10000 and cliente_2_activo == True: #Para comparar que una variable tiene cierto valo, usamos 2 iguales
  print("Es cliente VIP")
elif ingresos_2 >= 5000 or cliente_2_activo == True:
  print("Es cliente bueno")
else: print("El cliente es malo")
```
